```python
# get the framerate of the video
def read_framerate(video_path):
 
    video = cv2.VideoCapture(video_path)
    #train1_process/train01_384216_30.mp4
    #video = cv2.VideoCapture(" ../data/train1/train01.mp4")
    
    
    # Find OpenCV version
    (major_ver, minor_ver, subminor_ver) = (cv2.__version__).split('.')
 
    if int(major_ver)  < 3 :
        fps = video.get(cv2.cv.CV_CAP_PROP_FPS)
        print ("Frames per second using video.get(cv2.cv.CV_CAP_PROP_FPS): {0}".format(fps))
    else :
        fps = video.get(cv2.CAP_PROP_FPS)
        print ("Frames per second using video.get(cv2.CAP_PROP_FPS) : {0}".format(fps))
 
    video.release();
    return fps
```

```python
# convert video to frame
def read_frame(video_dir, imagesfile_dir):
    cap = cv2.VideoCapture(video_dir)

    # total frame number
    frames_num = cap.get(7)
    print("video1.frames_num: {}".format(frames_num))

    start_time = time.time()

    # one by one
    j = 0
    while(1):
        # read() retrun a boolean and a frame.  read successfully - rval: True
        rval, frame = cap.read()

        if rval is False:
            break

        image = os.path.join(imagesfile_dir, str(j) + ".jpg")
        print(image)
        cv2.imwrite(image, frame)
        j = j + 1

    cap.release()
    print('takes time {:.3f} seconds'.format(time.time() - start_time))
```
